We can remove the country column as all the rows have united states in their country column so the data set is independent of the country column

```python
df = df.drop('Country', axis=1)
```

CHECKING FOR NULL VALUES

```python
print(df.isnull().sum())
```

The Dataset doesn't contain any null values

```python
print(df.describe(include='object'))
```

```python
df['Category'].unique()
```

```python
category_counts = df['Category'].value_counts()
plt.figure(figsize=(5, 3))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Categories distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Sales'], linewidth=2)
plt.title('Distribution of Sales (KDE Line Plot)')
plt.xlabel('Sales')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=50, kde=True)
plt.title('Sales distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

We are going to find out all the cities included in the datset where the sales take place

```python
df['City'].unique()
```

The number of unique cities is very high so the graph that will be used to represent the sales there will be very complex

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Sales', y='City', data=df, ci=None)
plt.title('Distribution of Sales by City')
plt.xlabel('Sales')
plt.ylabel('City')
plt.show()
```

Plotting the Distribution of Sales by State

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Sales', y='State', data=df, ci=None)
plt.title('Distribution of Sales by State')
plt.xlabel('Sales')
plt.ylabel('State')
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
plt.figure(figsize=(8, 8))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Sales by Region (Pie Chart)')
plt.show()
```

Here we are going to use a pair plot to find connections  between the numerical data and check numbers and seeing a lot of data plotted together at once and choosing the important features and values and also to spot mistakes

```python
sns.pairplot(data=df, hue='Category')
plt.suptitle('Pairplot', y=1.02)
plt.show()
```

Now we are going to plot a Heat Map for multivariate analysis and for finding the correlation between different features of the dataset for analysis

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis')
plt.title('Correlation Heatmap')
plt.show()
```

The code below is to find the "Weak Spots" these are the spots or cities with the least amount of profits and we are going to plot the 10 cities with the least amount of Sales

```python
lowest_values_cities = 10
city_sales_mean = df.groupby('City')['Sales'].mean().reset_index()
bottom_cities = city_sales_mean.sort_values(by='Sales', ascending=True).head(lowest_values_cities)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Sales', y='City', data=bottom_cities, ci=None)
plt.title(f'Bottom {lowest_values_cities} Cities with the Least Mean sales')
plt.xlabel('Sales')
plt.ylabel('City')
plt.show()
```
